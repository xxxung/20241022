```python
import os
from dotenv import load_dotenv
load_dotenv()
#pip python-dotenv 虛擬網路變數
#創立 .env 放入密碼 (名稱=密碼)
#創立 .gitignore 不要上傳 (檔案名稱)
#電腦環境變數 保護資料 使用者看不到
key = os.environ['Gemini_API_KEY']  #使用網路變數

```

```python
#pip google-generativeai套件
import google.generativeai as genai

genai.configure(api_key=key)         #your api key
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content("如何學習python?,回應結果請輸出成為html格式")
print(response.text)
```
