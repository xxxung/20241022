```python
import pandas as pd 

df = pd.read_csv('上市公司資料.csv')
df1 = df.astype({'公司代號':'str'})
df2 = df1.drop(columns=['出表日期','資料年月','營業收入-去年當月營收','累計營業收入-當月累計營收','累計營業收入-去年累計營收','累計營業收入-前期比較增減(%)','備註'])


def color_negative(val):
    color = 'red' if float(val) < 0 else 'green'  # 判斷是否為負數
    return f'color: {color}'

df2.style\
.format({
        '營業收入-上月比較增減(%)': '{:.2f}%', 
        '營業收入-去年同月增減(%)': '{:.2f}%'
        },precision=2,thousands=',')\
.format_index(lambda v:v+1)\
.applymap(color_negative, subset =['營業收入-上月比較增減(%)'])


```
