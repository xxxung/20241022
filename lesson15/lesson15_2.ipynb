```python
import numpy as np 
import pandas as pd 

scores = np.random.randint(50,101,size=(50,5))
scores_df = pd.DataFrame(scores,index=range(1,51),columns=['國文','英文','數學','地理','歷史'])
scores_df.index.name = "學號"
scores_df.columns.name = "科目"
scores_df
```

## apply
## DataFrame.apply(function) 
- apply會將每個DataFrame裡的Series放入function參數中
- colums會是index

```python
def cal1(s):
    return s.mean()
scores_df.apply(cal1, axis=0)  #axis決定Series要橫的還是直的
```

```python
scores_df1 = scores_df.copy()
```

```python
def cal2(s):
    nums = int(s[s<60].count())
    return nums
scores_df.apply(cal2, axis=1)  #axis決定Series要橫的還是直的
scores_df1['不及格的科目']=scores_df.apply(cal2,axis=1)
scores_df1
```

```python
s1 = scores_df.iloc[0]
int(s1[s1<60].count())  #count計算數量
```

```python
def cal3(s):
    return pd.Series([s.min() , s.max() , s.std()],index=['最低分','最高分','標準差'])
scores_df3 = scores_df.apply(cal3)
scores_df3
```

## pd.concat([])  連結相同欄位的DataFrame

```python
pd.concat([scores_df,scores_df3])
```
