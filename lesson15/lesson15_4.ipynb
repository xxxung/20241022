## stack() 
- 將 DataFrame 中的每一列（如 Open, High, Low, Close, Adj Close 等）轉換成一個多層次索引（hierarchical index）中的值。每一列變成了 DataFrame 中的其中一個層級，對應於日期和股票的其他資料。
- 在 stack() 操作後，我們獲得的是一個多層次索引的 Series。接下來使用 ['Adj Close'] 來提取每個日期的 調整後收盤價（Adj Close）的數據。

```python
all_data['AAPL'].stack()['Adj Close']   #all_data是dict 
```

## hk

```python
import yfinance as yf
import pandas as pd

all_data = {ticker:yf.download(ticker) for ticker in ['2330.TW','2303.TW','2454.TW','2317.TW']}

new_data = {key:df[('Adj Close', key)] for key,df in all_data.items()}

df = pd.DataFrame(new_data)
df1 = df.rename(columns={
    '2330.TW':'台積電',
    '2303.TW':'聯電',
    '2454.TW':'聯發科',
    '2317.TW':'鴻海'
})
df1
```

```python
df2 = df1.dropna()
df2
```

```python
df2.info()
```

## pct_change() 與上一筆資料比較百分比

```python
df3 = df2.pct_change().dropna()
df3
```

## .corr() 相關係數

```python
df3.corr()
```

## .shift() 向下偏移

```python
(df2 - df2.shift(1)).dropna()
```
