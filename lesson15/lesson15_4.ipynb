```python
import yfinance as yf
import pandas as pd

all_data = {}
for ticker in ['2330.TW','2303.TW','2454.TW','2317.TW']:
    all_data[ticker] = yf.download(ticker)

# s1 = all_data['2330.TW'][('Adj Close','2330.TW')]
# s2 = all_data['2303.TW'][('Adj Close','2303.TW')]
# s3 = all_data['2454.TW'][('Adj Close','2454.TW')]
# s4 = all_data['2317.TW'][('Adj Close','2317.TW')]


new_data = {key:df[('Adj Close', key)] for key,df in all_data.items()}

df = pd.DataFrame(new_data)

df1 = df.rename(columns={
    '2330.TW':'台積電',
    '2303.TW':'聯電',
    '2454.TW':'聯發科',
    '2317.TW':'鴻海'
})
df1
```
