```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y,'ro-')

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1, 1, 1)

subject = ['chinese','english','math','history','discover']
nums1 = [1, 4, 9, 16, 25]
nums2 = [25,16,9,4,1]

axes.plot(subject,nums1,'o--g',label='student1')
axes.plot(subject,nums2,'+-r',label='student2') 

axes.set_title('students')           #title
axes.set_xlabel('subject')           #x軸標籤
axes.set_ylabel('scores')            #y軸標籤
axes.grid(axis='y')                  #格線(只有y軸線)

axes.set_xticks([0,1,2,3,4])
axes.set_xticklabels(['ch','en','ma','hi','cover'])

axes.legend()                        #線段說明,label加在axes.plot(後面)

plt.show()
```
