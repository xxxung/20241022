```python
import pandas as pd

df = pd.read_csv('全台火車進出站資訊.csv')
df
```

```python
df1 = df.groupby(by=['縣市','年度','站名'])[['進站人數','出戰人數']].max()
df1
```

```python
df1.columns.name = '人數'
s1 = df1.stack(level='人數')
s1
s1.unstack(level=['人數','年度'])
```
