## reindex() 查詢
## drop() 移除不要欄位

```python
#透過 reindex() 的選取

df.reindex(index=[0,1,2,3,1108,1109],columns=['證券名稱','收盤價','漲跌價差'])
```

```python
#透過 drop() '移除'欄位

df.drop(columns=['成交筆數','成交股數'])
```

## 更改欄位名稱

```python
#更改欄位名稱 , 使用 columns

df2 = df.reindex(index=[0,1,2,3,1108,1109],columns=['證券名稱','收盤價','漲跌價差'])
df2.columns = ['欄位1','欄位2','欄位3']
df2
```

```python
#更改欄位名稱 , 使用 rename()

df2 = df.reindex(index=[0,1,2,3,1108,1109],columns=['證券名稱','收盤價','漲跌價差'])
df2.rename(columns={'證券名稱':'證券','收盤價':'收盤'})
```
