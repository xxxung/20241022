```python
import pandas as pd 

df = pd.read_csv('world.csv')
df1 = df[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]  #留下需要的欄位
df1
```

## query()

```python
taiwan = df1.query('國家 == "台灣"')
taiwan
```

```python
_2020_s1 =taiwan.query('日期 >= "2020-01-16" and 日期 <= "2020-12-31" ')
_2020_s1
```

## 重要!

```python
country = input("請輸入國家:")
country_df = df1.query('國家 == @country')  #若 query 要搜尋的是一個變數 要加上@

_2021_country = country_df.query('日期 >= "2021-01-01" and 日期 <= "2021-12-31" ')
median = int(_2021_country['新增確診數'].median())
_2021_country.query('新增確診數 > @median')
```

```python
import pandas as pd

df = pd.read_csv('world.csv')
df1 = df[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]

country = input('請輸入國家:')
country_df = df1.query('國家 == @country')

max = int(country_df['新增確診數'].max())
country_df.query('新增確診數 == @max')
```
