## csv  (讀檔)
- next()
- csv.reader()


```python
import csv
with open('個股日成交資訊.csv',encoding='utf-8',newline='') as file:    #newline='' 保持原始樣子不換行輸出
    csv_reader = csv.reader(file)                                      #csv提供的reader(),得到一個reader的實體
    next(csv_reader)                                                   #跳過一個list:跳過檔案中第一列(欄位名稱)
    # print(type(csv_reader))
    for row in csv_reader:                                             #row是 list
        print(row)
```

## 這裡很重要  (讀取、過濾資料)
- csv.DictReader
- del

```python
import csv
with open('個股日成交資訊.csv',encoding='utf-8',newline='') as file:    #newline='' 保持原始樣子不換行輸出
    csv_dictReader = csv.DictReader(file,['證券代號','證券名稱','開盤價','收盤價','漲跌價差','成交筆數'],restkey='不要') #選擇需要的，其他的歸類成list為不要的
    next(csv_dictReader)
    csv_list:list[dict] = list(csv_dictReader)
    # print(type(csv_dictReader))                                        #實體
    # for row in csv_dictReader:
    #     print(row)                                                     #row是dict

for row in csv_list:
    del row['不要']     #刪除不要的欄位 del
    print(row)
```

## 存檔 (csv檔)
- csv.DictWriter
1. csv.DictWriter.writeheader()
2. csv.DictWriter.writerows()

```python
import csv
with open('new.csv',mode='w',encoding='utf-8',newline='') as file:
    csv_witer = csv.DictWriter(file,['證券代號','證券名稱','開盤價','收盤價','漲跌價差','成交筆數'])
    csv_witer.writeheader()
    csv_witer.writerows(csv_list)  #csv_list裡面一定要有list[dict]
```
