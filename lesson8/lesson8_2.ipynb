```python
import requests    #終端機pip install requests
from requests import Response
url ='https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=CSV'
r:Response = requests.get(url)
if r.ok:
    print('下載成功')
```

```python
from io import StringIO
import csv

data:str = r.text  
file = StringIO(data)    #StringIO(data) 會把傳入的字串 data 當作'檔案內容'，並返回一個可以讀取和寫入的類似檔案物件。
dictReader = csv.DictReader(file,fieldnames=['sitename','county','aqi','status','pm2.5','datacreationdate','longitude'],restkey='不要')
next(dictReader)
datas:list[dict] = list(dictReader)
for site in datas:
    del site['不要']
    print(site)
```

```python
with open('aqi.csv',mode='w',encoding='utf-8',newline='') as file:
    dict_writer = csv.DictWriter(file,fieldnames=['sitename','county','aqi','status','pm2.5','datacreationdate','longitude'])
    dict_writer.writeheader()
    dict_writer.writerows(datas)
```
