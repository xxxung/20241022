### pd.concat([DataFrame, DataFrame ])  裡面資料格式是list[DataFrame, DataFrame...]

```python
import os
import pandas as pd
from pandas import DataFrame

def merge_data(station_paths:list[str], info:str)->DataFrame:
    '''
    傳出所有的資料
    '''
    info_df = pd.read_csv(info)
    merge_datas = []
    for sta_path in station_paths:    #station_paths是一個list[str]
        station_df = pd.read_csv(sta_path)
        merge_df = pd.merge(station_df, info_df,how='left',left_on='staCode',right_on='stationCode')
        merge_datas.append(merge_df)
    return pd.concat(merge_datas)  # 合併成一個大DataFrame 並回傳


current_folder = os.getcwd()    #取得現在檔案路徑 str
stations_folder = os.path.join(current_folder,'stations')    #合併並取得stations絕對路徑 str
csv_abs_files = []
for filename in os.listdir(stations_folder):  #list[str]
    if '每日各站' in filename:
        csv_abs_files.append(os.path.join(stations_folder,filename))

info_path = os.path.join(stations_folder,'台鐵車站資訊.csv')  #取得 台鐵車站資訊.csv的絕對路徑  str


df = merge_data(csv_abs_files,info_path)
df1 = df[['trnOpDate','gateInComingCnt','gateOutGoingCnt','name','stationAddrTw','haveBike']]
```

## colums=[]

```python
df1.columns = ['日期','進站人數','出戰人數','站名','地址','Ubike']   #直接照順序改欄位名稱
df1
```

```python
df1.info()
```

## to_datetime()

```python
df1['日期'] = pd.to_datetime(df1['日期'].astype(str))
df1
```

## map()  裡面放dict直接改

```python
df1['Ubike'] = df1['Ubike'].map({'Y':1,'N':0})
```

```python
df1
```

## sort_values() 排序

```python
df2 = df1.sort_values(by='日期')
df2
```

## DataFrame.set_index() 欄位改為index

```python
df3 = df2.set_index('日期')
df3
```

```python
def first_third(val:str)->str:
    return val[:3]
df3['地址'].map(first_third)
```

```python
df3['縣市'] = df3['地址'].map(lambda v:v[:3])
```

## index.strftime('格式)

```python
df3["年度"] = df3.index.strftime('%Y')
df3["年月"] = df3.index.strftime('%Y-%m')
```

```python
df4 = df3.reindex(columns=['縣市','年度','年月','站名','進站人數','出戰人數','Ubike'])
df4
```

```python
df4.to_csv('全台火車進出站資訊.csv')
```
