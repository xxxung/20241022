```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())
_ = ax.set_xticks([0,100,250,350,500,600,750,850,1000])
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())

_ = ax.set_xticks([0,100,200,300,400,500,600,700,800,900,1000])
_ = ax.set_xticklabels(['zero','one','two','three','four','five','six','seven','eight','nine','ten'])
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())

_ = ax.set_xticks([0,100,200,300,400,500,600,700,800,900,1000])
_ = ax.set_xticklabels(['zero','one','two','three','four','five','six','seven','eight','nine','ten'])

#想要顯示的區域
ax.set_xlim(300,600) #顯示x 300-600
ax.set_ylim(-10,10)  #顯示y -10-10
```
