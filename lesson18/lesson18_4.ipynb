# DataFrame

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.randn(10,4).cumsum(axis=0),
             columns=['A','B','C','D'],
             index=np.arange(0,100,10))

df.plot.line()
```

```python
#Series
fig, axes = plt.subplots(2,1)

series = pd.Series(np.random.rand(16),
          index=list('abcdefghijklmnop'))

series.plot.bar(ax=axes[0], color='k', alpha=0.7)
series.plot.barh(ax=axes[1], color='k', alpha=0.7)   #barh 
```

## columnsIndex

```python
df = pd.DataFrame(np.random.rand(6,4),
             index=['one','two','three','four','five','six'],
             columns=pd.Index(['A','B','C','D'],name='Genus'))  #columnsIndex
df
```

```python
df.plot.bar()
```

```python
df.plot.barh(stacked=True, alpha=1)
```

## pd.crosstab 
## div 

```python
tips = pd.read_csv('tips.csv')

party_counts = pd.crosstab(tips['day'], tips['size'])  #
party_counts1 = party_counts.loc[:,2:5]                #loc 25
party_pcts = party_counts1.div(party_counts.sum(axis=1),axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```
