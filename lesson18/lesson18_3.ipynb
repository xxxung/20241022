# Series

## axes.annotate 圖表上標註
## set_title 設置圖形標題

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datetime import datetime

data = pd.read_csv('spx.csv',index_col=0, parse_dates=True)    #index_col=0 表示將第一列作為索引（通常是日期列）, parse_dates=True 解析日期格式
series = data['SPX']

fig = plt.figure(figsize=(10,5))
axes = fig.add_subplot(1,1,1)

crisis_data = [
    (datetime(2007, 10, 11),'Peak of bull market'),
    (datetime(2008, 3, 12),'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
#圖表上標註
for date, label in crisis_data:
    axes.annotate(text=label,                                                           #標註的名稱
                  xy=(date,series[date]+25),                                            #xy座標
                  xytext=(date,series[date]+225),                                       #設置文本的顯示位置（通常是箭頭的上方）。
                  arrowprops=dict(facecolor='red',headwidth=4,width=2,headlength=4),    #設置箭頭的顏色、寬度和頭部的大小。
                  horizontalalignment='left',                                           #顯示在箭頭的水平方向左或右
                  verticalalignment='top'                                               #顯示在箭頭垂直方向上或下
                  )

series.plot.line(ax=axes,style='k-')                   #直接用Series 畫圖  ax使用axes的子圖

_ = axes.set_title('Important dates in the 2008-2009 financial crisis')  #設置圖形標題
axes.get_xticks()

axes.set_xlim(datetime(2007,1,1),'2011-01-01')
axes.set_ylim([600, 1900])
```
