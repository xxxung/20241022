## randn 常態亂數
## sumsum 累計

```python
import numpy as np

fig = plt.figure()

ax1 = fig.add_subplot(2 ,2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
#plt.plot(np.random.randn(50).cumsum(),'k--')
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)    #直方圖, bins=分成 20 個區間, alpha=透明度 
ax2.scatter(np.arange(30), np.arange(30)+(3 * np.random.randn(30))) #散點圖 x, y 
ax3.plot(np.random.randn(50).cumsum(), 'k--')  #折線圖 

np.random.randn(10)
```

## plt.subplots    創建圖形和子圖
## subplots_adjust 調整子圖距離

```python
#同時建立figure,axes
#top funciton -> subplots
fig, axes = plt.subplots(2,2,sharex=True,sharey=True) #2列2欄的圖, sharex=True,sharey=True:共享相同的 x 軸和 y 軸

#左上角方格為[0,0] 右上角[0,1] 左下角[1,0] 右下角[1,1]
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5) #直方圖

fig.subplots_adjust(wspace=0,hspace=0)  #調整子圖距離 wspace=0：設置水平方向上的間距為 0 ,  hspace=0：設置垂直方向上的間距為 0

print(axes)
```

## plt.plot
- label 線的圖例名稱
- drawstyle  畫的樣式
## plt.legend 顯示圖利
- loc='best' 自動選擇最佳的位置來放置圖例，避免與圖形重疊。 

```python
data = np.random.randn(30).cumsum()

plt.plot(data,'k--',label='Default')   #label:線的圖例名稱為 "Default"
plt.plot(data,'k-',drawstyle='steps-post',label='steps-post')

plt.legend(loc='best')
```
